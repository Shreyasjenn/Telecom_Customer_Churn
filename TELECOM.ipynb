# QUESTIONS

###Descriptive Statistics

# 1)Which major 3 cities have more no. of customers ?

```python
city_cust_count=df['City'].value_counts()
top_3=city_cust_count.head(3)
top_3
```

```python
top_3.plot(kind='bar', color='skyblue')
plt.title('Top 3 Cities with Most Customers')
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.show()
```


# 2. How many people are using credit cards for payment purpose?

```python
num_credit_card_users = df['Payment Method'].value_counts().get('Credit Card', 0)
print("Number of people using credit cards for payment:", num_credit_card_users)
```

# 3)How many male and female customers the company has ?

```python
gender_counts = df['Gender'].value_counts()
print(gender_counts)
gender_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Gender Distribution')
plt.show()
```

# 4)What is the number of customers having dependents ?

```python
dependents_counts = df['Number of Dependents'].value_counts()
print(dependents_counts)
dependents_counts.plot(kind='bar', color='lightgreen')
plt.title('Customers with Dependents')
plt.xlabel('Dependents')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.show()
```

# 5)How many customers used internet services?

```python
internet_service_counts = df['Internet Service'].value_counts()
print(internet_service_counts)
internet_service_counts.plot(kind='line', marker='o', color='skyblue')
plt.title('Internet Service Usage')
plt.xlabel('Internet Service')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.show()
```

# 6)Give the customer city distribution

```python
city_counts= df['City'].value_counts().head(10)
print()
city_counts.plot(kind='line', color='skyblue')
plt.title('Customer City Distribution')
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.xticks(rotation=30)
```

# 7)Average Monthly GB Download 

```python
plt.boxplot(df['Avg Monthly GB Download'])
plt.title('Monthly GB Download Distribution')
plt.ylabel('Monthly GB Download')
plt.show()
```

# 8)How many customers have churned and how many stayed ?

```python
sns.histplot(data=df, x='Customer Status',color='Blue')
```

# 9)Which customer has been using the services for the longest period?

```python
customer_longest_tenure = df.loc[df['Tenure in Months'].idxmax()]['Customer ID']
print("Customer with the longest tenure:", customer_longest_tenure)
```

# 10)Which customer pays the most bill?

```python
customer_highest_bill = df.loc[df['Monthly Charge'].idxmax()]['Customer ID']
print("Customer who pays the most bill:", customer_highest_bill)
```

###Payment Analysis

# 11)Which payment methods are available and how many customers are using those ?

```python
major_payment_methods = df['Payment Method'].value_counts().head(3)
plt.figure(figsize=(8, 6))
plt.pie(major_payment_methods, labels=major_payment_methods.index, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Major Payment Methods Used')
plt.show()
```

# 12)How many customers are not using Premium Tech Support?

```python
premium_tech_support_counts = df['Premium Tech Support'].value_counts()
premium_tech_support_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Premium Tech Support Usage')
plt.show()
```

###Referral Analysis

# 13)List of customers who have referred others to this company 10 or more than 10 time ?

```python
referral_customers = df[df['Number of Referrals'] >= 10]['Customer ID'].head(10)
print("Customers who referred others 10 or more times:\n", referral_customers)
```

```python
referral_counts = df['Number of Referrals'].value_counts()
referral_counts.plot(kind='bar', color='skyblue')
plt.title('Referral Count Distribution')
plt.xlabel('Number of Referrals')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.show()
```

###Churn Analysis

# 14)What is the most common reason to churn?

```python
churn_reason_counts = df['Churn Reason'].value_counts().head(10)
plt.figure(figsize=(15,6))
churn_reason_counts.plot(kind='bar', color='lightcoral')
plt.title('Churn Reasons Distribution')
plt.xlabel('Churn Reason')
plt.ylabel('Number of Customers')
plt.show()
```

# 15)Number of customers to whom the offers are allotted?

```python
num_customers_with_offers = df['Offer'].notnull().sum()
print("Number of customers to whom offers are allotted:", num_customers_with_offers)
```

```python
offer_counts = df['Offer'].value_counts()
offer_counts.plot(kind='line', color='lightgreen')
plt.title('Offer Distribution')
plt.xlabel('Offer')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```

# 16) Which type of offer is most commonly allotted to customers?

```python
common_offer_type = df['Offer'].value_counts().idxmin()
print("Most common offer type allotted to customers:", common_offer_type)
```

###Usage Analysis

# 17)Which internet types are most commonly used?

```python
internet_service_types_counts= df['Internet Type'].value_counts()
internet_service_types_counts.plot(kind='bar', color=['lightgreen','lightblue','lightpink','orange'])
plt.title('Internet Service Types Distribution')
plt.ylabel('Number of Customers')
plt.xticks(rotation=0)
plt.show()
```

# 18)Which customers use multiple lines service?

```python
customers_with_multiple_lines = df[df['Multiple Lines'] == 'Yes']['Customer ID'].head(5)
print("Customers using multiple lines service:\n", customers_with_multiple_lines)
```

```python
multiple_lines_counts = df['Multiple Lines'].value_counts()
multiple_lines_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Multiple Lines Usage')
plt.show()
```
